```python
file_STD_mean = '/Users/annasommer/Documents/Documents/OSSE_Project/Data_for_plot/STD_mean_TD.txt'
file_Corr_mean = '/Users/annasommer/Documents/Documents/OSSE_Project/Data_for_plot/Corr_mean_TD.txt'
file_RMSE_mean = '/Users/annasommer/Documents/Documents/OSSE_Project/Data_for_plot/RMSE_mean_TD.txt'
file_RMSE_cent_mean = '/Users/annasommer/Documents/Documents/OSSE_Project/Data_for_plot/RMSE_cent_mean_TD.txt'

with open(file_STD_mean) as f:
        rows = [[str(x) for x in line.split(';')] for line in f]
        biome_mean_STD,model_mean_STD,STD_mean = [list(col) for col in zip(*rows)] 
        f.close()
        
with open(file_Corr_mean) as f:
        rows = [[str(x) for x in line.split(';')] for line in f]
        biome_mean,model_mean,Corr_mean = [list(col) for col in zip(*rows)] 
        f.close()

with open(file_RMSE_mean) as f:
        rows = [[str(x) for x in line.split(';')] for line in f]
        biome_mean,model_mean,RMSE_mean = [list(col) for col in zip(*rows)] 
        f.close()

with open(file_RMSE_cent_mean) as f:
        rows = [[str(x) for x in line.split(';')] for line in f]
        biome_mean,model_mean,RMSE_cent_mean = [list(col) for col in zip(*rows)] 
        f.close()
        
file_STD_all = '/Users/annasommer/Documents/Documents/OSSE_Project/Data_for_plot/STD_all_TD.txt'
file_Corr_all = '/Users/annasommer/Documents/Documents/OSSE_Project/Data_for_plot/Corr_all_TD.txt'
file_RMSE_all = '/Users/annasommer/Documents/Documents/OSSE_Project/Data_for_plot/RMSE_all_TD.txt'
file_RMSE_cent_all = '/Users/annasommer/Documents/Documents/OSSE_Project/Data_for_plot/RMSE_cent_all_TD.txt'

with open(file_STD_all) as f:
        rows = [[str(x) for x in line.split(';')] for line in f]
        biome_all,model_all,k_numb,STD_all = [list(col) for col in zip(*rows)] 
        f.close()
        
with open(file_Corr_all) as f:
        rows = [[str(x) for x in line.split(';')] for line in f]
        biome_all,model_all,k_numb,Corr_all = [list(col) for col in zip(*rows)] 
        f.close()

with open(file_RMSE_all) as f:
        rows = [[str(x) for x in line.split(';')] for line in f]
        biome_all,model_all,k_numb,RMSE_all = [list(col) for col in zip(*rows)] 
        f.close()

with open(file_RMSE_cent_all) as f:
        rows = [[str(x) for x in line.split(';')] for line in f]
        biome_all,model_all,k_numb,RMSE_cent_all = [list(col) for col in zip(*rows)] 
        f.close()        
        
biome_mean_STD = np.array(biome_mean_STD)
biome_mean_STD = biome_mean_STD.astype(np.float)
model_mean_STD = np.array(model_mean_STD)
model_mean_STD = model_mean_STD.astype(np.float)
biome_mean = np.array(biome_mean)
biome_mean = biome_mean.astype(np.float)
model_mean = np.array(model_mean)
model_mean = model_mean.astype(np.float)

STD_mean = np.array(STD_mean)
STD_mean = STD_mean.astype(np.float)
Corr_mean = np.array(Corr_mean)
Corr_mean = Corr_mean.astype(np.float)
RMSE_mean = np.array(RMSE_mean)
RMSE_mean = RMSE_mean.astype(np.float)
RMSE_cent_mean = np.array(RMSE_cent_mean)
RMSE_cent_mean = RMSE_cent_mean.astype(np.float)

biome_all = np.array(biome_all)
biome_all = biome_all.astype(np.float)
model_all = np.array(model_all)
model_all = model_all.astype(np.float)
k_numb = np.array(k_numb)
k_numb = k_numb.astype(np.float)

STD_all = np.array(STD_all)
STD_all = STD_all.astype(np.float)
Corr_all = np.array(Corr_all)
Corr_all = Corr_all.astype(np.float)
RMSE_all = np.array(RMSE_all)
RMSE_all = RMSE_all.astype(np.float)
RMSE_cent_all = np.array(RMSE_cent_all)
RMSE_cent_all = RMSE_cent_all.astype(np.float)
```

```python
biome_numb = [0,9,10,11,12,13,15,16,17]
```

```python
STD_mean_tot = np.zeros((12,9))
Corr_mean_tot = np.zeros((12,9))
RMSE_mean_tot = np.zeros((12,9))
RMSE_cent_mean_tot = np.zeros((12,9))

STD_all_tot = np.zeros((12,4,9))
Corr_all_tot = np.zeros((12,4,9))
RMSE_all_tot = np.zeros((12,4,9))
RMSE_cent_all_tot = np.zeros((12,4,9))

for i in np.arange(0,12,1):
    p = 0
    for j in biome_numb:
        p = p + 1
        for k in np.arange(0,len(STD_mean),1):
            if i == int(model_mean_STD[k]) and j == int(biome_mean_STD[k]):
                STD_mean_tot[i,p-1] = STD_mean[k]
                break
                
for i in np.arange(0,12,1):
    p = 0
    for j in biome_numb:
        p = p + 1
        for k in np.arange(0,len(Corr_mean),1):
            if i == 0:
                Corr_mean_tot[i,p-1] = 1.
                RMSE_mean_tot[i,p-1] = 0.
                RMSE_cent_mean_tot[i,p-1] = 0.
                break 
            if i == int(model_mean[k]) and j == int(biome_mean[k]):                
                Corr_mean_tot[i,p-1] = Corr_mean[k]
                RMSE_mean_tot[i,p-1] = RMSE_mean[k]
                RMSE_cent_mean_tot[i,p-1] = RMSE_cent_mean[k]
                break

for i in np.arange(0,12,1):
    p = 0
    for j in biome_numb:
        p = p + 1
        for n in np.arange(0,4,1):
            for k in np.arange(0,len(Corr_mean),1):
                if i == 0:
                    STD_all_tot[i,n,p-1] = STD_mean_tot[0,p-1]
                    Corr_all_tot[i,n,p-1] = 1.
                    RMSE_all_tot[i,n,p-1] = 0.
                    RMSE_cent_all_tot[i,n,p-1] = 0.
                    break 
                if i == int(model_mean[k]) and j == int(biome_mean[k]) and n == int(k_numb[k]):   
                    STD_all_tot[i,n,p-1] = STD_all[k]
                    Corr_all_tot[i,n,p-1] = Corr_all[k]
                    RMSE_all_tot[i,n,p-1] = RMSE_all[k]
                    RMSE_cent_all_tot[i,n,p-1] = RMSE_cent_all[k]
                    break
```
