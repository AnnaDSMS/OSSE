```python
def readnc_1d(ncfile,varname):
        fid = nc.netcdf_file(ncfile, 'r')
        out = np.array(fid.variables[varname][:]).squeeze()
        fid.close()
        return out
```

```python
def readnc_2d(ncfile,varname):
        fid = nc.netcdf_file(ncfile, 'r')
        out = np.array(fid.variables[varname][:,:]).squeeze()
        fid.close()
        return out
```
